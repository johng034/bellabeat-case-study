```python
def create_date(date):
    date_list = date.split('/')
    month, day, year = date_list

    if len(month) != 2:
        month = '0' + month
    if len(day) != 2:
        day = '0'+ day

    corrected_date = '-'.join([month, day, year])

    return corrected_date
```

```python
def fix_date(df, column):
    temp_df = df.copy()

    temp_df['temp_date'] = temp_df[column].apply(lambda x: x[:9].strip())

    corrected_dates = temp_df['temp_date'].apply(lambda x: create_date(x))

    df[column] = corrected_dates

    return df
```

# Fix the Daily Datasets
- Convert the timestamps in a form that SQL will accept
    - Example: '4/12/2016' $\rightarrow$ '04-12-2016'

```python
temp_sleep = sleep_df.copy()
new_sleep_df = fix_date(temp_sleep, 'SleepDay')

temp_weight = weight_df.copy()
new_weight_df = fix_date(temp_weight, 'Date')
```

## Join the Dataframes

```python
# Merge calorie and intensity dataframes
cal_intensity = hourly_cal.merge(hourly_intensity, how='outer', on=['Id', 'ActivityHour'])

# Merge the previous dataframe with the steps dataframe
merged_df = cal_intensity.merge(hourly_steps, how='outer', on=['Id', 'ActivityHour'])

merged_df.head()
```

```python
# Create a new column with the time
merged_df['Time'] = merged_df['ActivityHour'].apply(lambda x: x[-11:].strip())
```

```python
# Clean the up the dates
new_merged_df = fix_date(merged_df, 'ActivityHour').copy()

# Reorder columns
column_names = ["Id", "ActivityHour", "Time", "StepTotal", "Calories", "TotalIntensity", "AverageIntensity"]
new_merged_df = new_merged_df.reindex(columns=column_names)

new_merged_df.head()
```

```python
new_merged_df['FullDate'] = new_merged_df['ActivityHour'] + ' ' + new_merged_df['Time']
```

```python
new_merged_df.head()
```

```python
# Convert datetime from 
# MM-DD-YYYY to YYYY-MM-DD

def reorder_date(date):
    month, day, year = date.split('-')
    return '-'.join([year, month, day])


# def convert_date(df, column):
#     fixed_date = df[column].apply(lambda x: reorder_date(x))
#     return fixed_date
```

```python
new_merged_df['FullDate'] = new_merged_df['ActivityHour'].apply(lambda x: reorder_date(x))

new_merged_df.head()
```

```python
def convert_time(fulltime):
    am_pm = fulltime[-2:]
    
    # Morning
    if am_pm == 'AM':
        time = fulltime[:-2]
        
    # Evening
    elif am_pm == 'PM':
        # Get the hour
        hour = fulltime[:2].replace(':', '')
        
        # Add 12 to convert to 24-hour time
        new_hour = int(hour) + 12
        
        if len(str(hour)) == 2:
            time = str(new_hour) + fulltime[2:-2]
        elif len(str(hour)) == 1:
            time = str(new_hour) + fulltime[1:-2]
    
    # Fix noon and midnight
    if time[:2] == '12':
        time = '00' + time[2:]
    elif time[:2] == '24':
        time= '12' + time[2:]
    
    return time
```

```python
new_merged_df['FullTime'] = new_merged_df['Time'].apply(lambda x: convert_time(x))
```

```python
# Combine the date and time columns
new_merged_df['FullDatetime'] = new_merged_df['FullDate'] + ' ' + new_merged_df['FullTime']
```

```python
final_df = new_merged_df[['Id', 'FullDatetime', 'StepTotal', 'Calories', 'TotalIntensity', 'AverageIntensity']].copy()

final_df.head()
```
